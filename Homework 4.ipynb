```python
import matplotlib.pyplot as plt
marks = {'Andy':88, 'Amy':66, 'James':90, 'Jules':55, 'Arthur':77}

plt.bar(marks.keys(), marks.values())
plt.xlabel('Student')
plt.ylabel('Grade')
plt.title('Bar Chart of Student Grades')

plt.show()
```

```python
plt.figure()
plt.pie(marks.values(), labels=marks.keys())
plt.title('Pie Chart of Student Grades')
plt.legend(title='Students', loc='best')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 6)
y1 = np.array([1, 3, 2, 4, 3])
y2 = np.array([3, 1, 4, 2, 5])
labels = ['A', 'B', 'C', 'D', 'E']
sizes = np.array([20, 30, 10, 15, 25])
colors = ['red', 'green', 'blue', 'yellow', 'orange']
x_grid = np.linspace(0, 5, 100)
y_grid = np.sin(x_grid)

fig, axs = plt.subplots(2, 3, figsize=(10, 6))

axs[0, 0].plot(x, y1, 'r-', label='Line 1')
axs[0, 0].plot(x, y2, 'g--', label='Line 2')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')
axs[0, 0].set_title('Multiple Lines')
axs[0, 0].legend()

names = ['Alice', 'Bob', 'Charlie', 'David', 'Eva']
values = [5, 3, 7, 2, 9]
axs[0, 1].bar(names, values)
axs[0, 1].set_title('Bar Chart')

labels = ['A', 'B', 'C', 'D', 'E']
sizes = [15, 30, 20, 10, 25]
axs[0, 2].pie(sizes, labels=labels, autopct='%1.1f%%')
axs[0, 2].set_title('Pie Chart')

axs[1, 0].plot(x_grid, y_grid)
axs[1, 0].grid()
axs[1, 0].set_xlabel('X-axis')
axs[1, 0].set_ylabel('Y-axis')
axs[1, 0].set_title('Grid Plot')

axs[1, 1].scatter(x, y1)
axs[1, 1].set_xlabel('X-axis')
axs[1, 1].set_ylabel('Y-axis')
axs[1, 1].set_title('Scatter Plot')

axs[1, 2].hist(y1, bins=5)
axs[1, 2].set_xlabel('Value')
axs[1, 2].set_ylabel('Frequency')
axs[1, 2].set_title('Histogram Plot')

plt.tight_layout()
plt.show()
```
