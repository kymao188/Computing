```python
import turtle as t 
```

```python
# flag(h) defines the outline of the flag- it is based on the ratios of the height (h) to the width, so it can be any scale you want
def flag(h):
    t.setx(0)
    t.sety(0)
    t.fd(1.9*h)
    t.lt(90)
    t.fd(h)
    t.lt(90)
    t.fd(1.9*h)
    t.lt(90)
    t.fd(h)
```

```python
# hoist(h,c) defines the box in the top left. Because the box is blue, we need a color parameter in addition to the height and a "fill" line
def hoist(h,c):
#t.fillcolor set the color for inside the hoist box
    t.fillcolor(c)
    t.lt(180)
    t.fd(h)
#these steps returns the turtle to (0,0) by running back up the left edge of the flag
    t.begin_fill()
    t.rt(90)
    t.fd(.76*h)
    t.rt(90)
    t.fd(.5385*h)
    t.rt(90)
    t.fd(.76*h)
    t.end_fill()
    t.lt(90)
    t.fd(.4615*h)
    t.lt(90)
```

```python
#the stripes are created based on color and measurements proportional to the realistic measurements of the american flag.
#13 stripes alternating from red to white.
#each stripe was defined as it's own based on it's position and color.
def stripe1(h,c2):  
    t.begin_fill()
    t.fillcolor(c2)
    t.fd(1.9*h)
    t.lt(90)
    t.fd(.0769*h)
    t.rt(90)
    t.fd(-1.9*h)
    t.rt(90)
    t.fd(.0769*h)
    t.end_fill()
    t.fd(-.0769*h)
    t.lt(90)
def stripes1(h,c1,c2):
    stripe1(h,c1)
    stripe1(h,c2)
def stripes1final (h,c1,c2):
    stripes1(h, c1,c2)
    stripes1(h, c1,c2)
    stripes1(h, c1,c2)
    t.fd(.76*h)
def stripe2 (h,c2):
    t.begin_fill()
    t.fillcolor(c2)
    t.fd(1.14*h)
    t.lt(90)
    t.fd(.0769*h)
    t.rt(90)
    t.fd(-1.14*h)
    t.rt(90)
    t.fd(.0769*h)
    t.end_fill()
    t.fd(-.0769*h)
    t.lt(90)
def stripes2(h,c1,c2):
    stripe2(h,c1)
    stripe2(h,c2)
def stripes2final (h,c1,c2):
    stripes2(h, c1,c2)
    stripes2(h, c1,c2)
    stripes2(h, c1,c2)
    stripe2(h,c1)
    t.end_fill()
def stripes(h,c,c1, c2):
    stripes1final(h,c1,c2)
    stripes2final(h,c1,c2)
    t.rt(180)
    t.fd(.76*h)
    t.lt(90)
    t.fd(.07*h)
    t.color(c)
    t.lt(90)
    t.fd(.1*h)
```

```python
#stars were drawn in 5 rows of 6 stars along with 4 rows of 5 stars, making 50 stars in total.
def star(h,c,c2):
    t.color(c2)
    t.fillcolor(c2)
    stariterator=(1,2,3,4,5)
    for x in stariterator:
        t.begin_fill()
        t.rt(144)
        t.fd(.05*h)
        t.end_fill()
#stars were first coded, then coded into rows accordingly.
def row1 (h,c,c2):
    star(h,c,c2)
    row1it=(1,2,3,4,5)
    for x in row1it:
        t.color(c)
        t.fd(.12*h)
        star(h,c,c2)
def row1s (h,c,c2):
    row1(h,c,c2)
    iterator=(1,2,3,4)
    for x in iterator:
        t.penup()
        t.fd(-.6*h)
        t.rt(90)
        t.fd(.1*h)
        t.lt(90)
        t.down()
        row1(h,c,c2)
    t.penup()
    t.setheading(90)
    t.fd(.35*h)
    t.lt(90)
    t.fd(.55*h)
    t.pendown()
    t.setheading(0)
def row2 (h,c,c2):
    star(h,c,c2)
    row1it=(1,2,3,4)
    for x in row1it:
        t.color(c)
        t.fd(.12*h)
        star(h,c,c2)
def row2s (h,c,c2):
    row2(h,c,c2)
    iterator=(1,2,3)
    for x in iterator:
        t.penup()
        t.fd(-.48*h)
        t.rt(90)
        t.fd(.1*h)
        t.lt(90)
        t.down()
        row2(h,c,c2)
```

```python
#the final step was to set colors to each corresponding letter (c/c1/c2) and define the main function.
def main (h,c,c1,c2):
    t.speed(50000)
    flag(h)
    hoist (h, c)
    stripes(h,c,c1,c2)
    row1s(h,c,c2)
    row2s (h,c,c2)
    t.ht()
    t.up()
    t.fd(h)
if __name__ == "__main__":
    main(300, "blue", "red", "white")
```
